```python
from sklearn.preprocessing import MinMaxScaler

# 피처값 스케일링
scaler = MinMaxScaler()
df1[['open','high','low','volume']] = scaler.fit_transform(df1[['open','high','low','volume']])
df2[['open','high','low','volume']] = scaler.fit_transform(df2[['open','high','low','volume']])
df3[['open','high','low','volume']] = scaler.fit_transform(df3[['open','high','low','volume']])
df4[['open','high','low','volume']] = scaler.fit_transform(df4[['open','high','low','volume']])

# 타켓 스케일링
scaler1 = MinMaxScaler()
df1['close'] = scaler1.fit_transform(df1['close'].values.reshape(-1,1))
df2['close'] = scaler1.fit_transform(df2['close'].values.reshape(-1,1))
df3['close'] = scaler1.fit_transform(df3['close'].values.reshape(-1,1))
df4['close'] = scaler1.fit_transform(df4['close'].values.reshape(-1,1))


# 스케일링 결과 확인
df1
```

```python
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(df1.drop('close', 1), df1['close'], test_size=0.2, random_state=11, shuffle=False)
x2_train, x2_test, y2_train, y2_test = train_test_split(df2.drop('close', 1), df2['close'], test_size=0.2, random_state=11, shuffle=False)
x3_train, x3_test, y3_train, y3_test = train_test_split(df3.drop('close', 1), df3['close'], test_size=0.2, random_state=11, shuffle=False)
x4_train, x4_test, y4_train, y4_test = train_test_split(df4.drop('close', 1), df4['close'], test_size=0.2, random_state=11, shuffle=False)
```

```python
print('train set 확인: ' ,  x1_train.shape, y1_train.shape, 'test set 확인: ' ,  x1_test.shape, y1_test.shape)
print('train set 확인: ' ,  x2_train.shape, y2_train.shape, 'test set 확인: ' ,  x2_test.shape, y2_test.shape)
print('train set 확인: ' ,  x3_train.shape, y3_train.shape, 'test set 확인: ' ,  x3_test.shape, y3_test.shape)
print('train set 확인: ' ,  x4_train.shape, y4_train.shape, 'test set 확인: ' ,  x4_test.shape, y4_test.shape)
```

```python
import tensorflow as tf
def windowed_dataset(series, window_size, batch_size, shuffle):
    series = tf.expand_dims(series, axis=-1)
    ds = tf.data.Dataset.from_tensor_slices(series)
    ds = ds.window(window_size + 1, shift=1, drop_remainder=True)
    ds = ds.flat_map(lambda w: w.batch(window_size + 1))
    if shuffle:
        ds = ds.shuffle(1000)
    ds = ds.map(lambda w: (w[:-1], w[-1]))
    return ds.batch(batch_size).prefetch(1)
```

```python
WINDOW_SIZE=20
BATCH_SIZE=32
```

```python
train1_data = windowed_dataset(y1_train, WINDOW_SIZE, BATCH_SIZE, True)
test1_data = windowed_dataset(y1_test, WINDOW_SIZE, BATCH_SIZE, False)

train2_data = windowed_dataset(y2_train, WINDOW_SIZE, BATCH_SIZE, True)
test2_data = windowed_dataset(y2_test, WINDOW_SIZE, BATCH_SIZE, False)

train3_data = windowed_dataset(y3_train, WINDOW_SIZE, BATCH_SIZE, True)
test3_data = windowed_dataset(y3_test, WINDOW_SIZE, BATCH_SIZE, False)

train4_data = windowed_dataset(y4_train, WINDOW_SIZE, BATCH_SIZE, True)
test4_data = windowed_dataset(y4_test, WINDOW_SIZE, BATCH_SIZE, False)
```

```python
for data in train1_data.take(1):
    print(f'데이터셋(X) 구성(batch_size, window_size, feature갯수): {data[0].shape}')
    print(f'데이터셋(Y) 구성(batch_size, window_size, feature갯수): {data[1].shape}\n')

for data in train2_data.take(1):
    print(f'데이터셋(X) 구성(batch_size, window_size, feature갯수): {data[0].shape}')
    print(f'데이터셋(Y) 구성(batch_size, window_size, feature갯수): {data[1].shape}\n')    
    
for data in train3_data.take(1):
    print(f'데이터셋(X) 구성(batch_size, window_size, feature갯수): {data[0].shape}')
    print(f'데이터셋(Y) 구성(batch_size, window_size, feature갯수): {data[1].shape}\n')
    
for data in train4_data.take(1):
    print(f'데이터셋(X) 구성(batch_size, window_size, feature갯수): {data[0].shape}')
    print(f'데이터셋(Y) 구성(batch_size, window_size, feature갯수): {data[1].shape}\n')
    
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv1D, Lambda
from tensorflow.keras.losses import Huber
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint


model = Sequential([
    # 1차원 feature map 생성
    Conv1D(filters=32, kernel_size=5,
           padding="causal",
           activation="relu",
           input_shape=[WINDOW_SIZE, 1]),
    # LSTM
    LSTM(16, activation='tanh'),
    Dense(16, activation="relu"),
    Dense(1),
])
```

```python
loss = Huber()
optimizer = Adam(0.0005)
model.compile(loss=Huber(), optimizer=optimizer, metrics=['mse'])
```

```python
earlystopping = EarlyStopping(monitor='val_loss', patience=10) 
```

```python
history1 = model.fit(train1_data, 
                    validation_data=(test1_data), 
                    epochs=50, 
                    callbacks=[earlystopping])

history2 = model.fit(train2_data, 
                    validation_data=(test2_data), 
                    epochs=50, 
                    callbacks=[earlystopping])

history3 = model.fit(train3_data, 
                    validation_data=(test3_data), 
                    epochs=50, 
                    callbacks=[earlystopping])

history4 = model.fit(train4_data, 
                    validation_data=(test4_data), 
                    epochs=50, 
                    callbacks=[earlystopping])
```

```python
pred1 = model.predict(test1_data)
pred2 = model.predict(test2_data)
pred3 = model.predict(test3_data)
pred4 = model.predict(test4_data)
```
